# JOIN


```python
# Left   Join

# Right  Join
# inner   Join
# Outer  Join
```

```python
import pandas as pd
one  =  pd.DataFrame({
            'Name':['Alex','Akau','Alice','Ayoung','Allen'],
            'Subject_id':['sub1','sub2','sub4','sub6','sub5'],
            'Mark_scored':[98,90,87,69,78]},
            index=[1,2,3,4,5])
one
```

```python
import pandas as pd
two  =  pd.DataFrame({
            'Name':['Billy','Brain','Brayce','Betty','Brucee'],
            'Subject_id':['sub2','sub4','sub3','sub6','sub5'],
            'Mark_scored':[89,80,79,97,88]},
            index=[1,2,3,4,5])
two
```

```python
print(pd.merge(one,two, on='Subject_id', how='left'))
```

```python
print(pd.merge(one,two, on='Subject_id', how='right'))
```

```python
print(pd.merge(one,two, on='Subject_id', how='inner'))
```

```python
print(pd.merge(one,two, on='Subject_id', how='outer'))
```
