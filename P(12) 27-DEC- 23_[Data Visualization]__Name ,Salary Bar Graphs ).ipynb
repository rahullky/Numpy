```python
import matplotlib.pyplot as plt
def  add(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
fig=plt.figure(figsize=(15,5))
plt.bar(n2,n3,color='mistyrose')
plt.xlabel('Name',size=20)         
plt.ylabel('Salary',size=20)
plt.title("Salary wise Data ",size=30,color='brown')
add(n2,n3)

plt.show()
# Xlabel  increase / decrease..size  ..color 
# Ylabel  increase / decrease..size  ..color 
# Title   add karna 
# def Function : ...salary ke uper salary show kare 
```

```python
fig=plt.figure(figsize=(15,5))
plt.bar(n2,n3,color='mistyrose')
plt.xlabel('Name',size=20)         
plt.ylabel('Salary',size=20)
plt.xticks(rotation=30,color='purple')
plt.yticks(rotation=30,color='purple')         
plt.title("Salary wise Data ",size=30,color='brown')
add(n2,n3)

plt.show()
# Xlabel  increase / decrease..size  ..color 
# Ylabel  increase / decrease..size  ..color 
# Title   add karna 
# def Function : ...salary ke uper salary show kare 
# Rotation   ...karana Xlabel ,Ylabel 
```
