```python
import matplotlib.pyplot as plt
def  add(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
fig=plt.figure(figsize=(15,5))
plt.bar(n2,n3,color='mistyrose')
plt.xlabel('Name',size=20)         
plt.ylabel('Salary',size=20)
plt.title("Salary wise Data ",size=30,color='brown')
add(n2,n3)

plt.show()
# Xlabel  increase / decrease..size  ..color 
# Ylabel  increase / decrease..size  ..color 
# Title   add karna 
# def Function : ...salary ke uper salary show kare 
```

```python
fig=plt.figure(figsize=(15,5))
plt.bar(n2,n3,color='mistyrose')
plt.xlabel('Name',size=20)         
plt.ylabel('Salary',size=20)
plt.xticks(rotation=30,color='purple')
plt.yticks(rotation=30,color='purple')         
plt.title("Salary wise Data ",size=30,color='brown')
add(n2,n3)

plt.show()
# Xlabel  increase / decrease..size  ..color 
# Ylabel  increase / decrease..size  ..color 
# Title   add karna 
# def Function : ...salary ke uper salary show kare 
# Rotation   ...karana Xlabel ,Ylabel 
```

# Mutiple Slicing  Pie Chart Me  (Aman,bunty,harsh) bhar nikalo

```python
import matplotlib.pyplot as plt
exp=(0,0,0,0.5,0,0,0,0.5,0,0,0,0.5)   # Col ki Lenght ==Len   hona chaiye 
                                     # 0.5  se row/colomn  slicing ho jayega 
```

```python
fig=plt.figure(figsize=(10,10))
plt.pie(n3,explode=exp, labels=n2,shadow=True)   
plt.show()
# Pie me Shadow ki  parchai   increase/decrease
```

# Slicing (% Percentage ) Pie Chart Me  (Aman,bunty,harsh) bhar nikalo


```python
fig=plt.figure(figsize=(15,10))
plt.pie(n3,explode=exp,labels=n2,shadow=True,autopct='%1.2f%%')
plt.show()
```
